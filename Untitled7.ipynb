```python
import numpy as np
import matplotlib.pyplot as plt

g = 9.8
a = 0.0
b = 10.0
N = 100
Np = N+1
h = (b-a)/N
target = 1e-10
x_values = [50, 100, 200, 500, 1000, 2000]

for idx, x_val in enumerate(x_values, 1):
    x = x_val * np.ones(Np, float)
    x[0] = 0
    x[N] = 0
    xp = np.zeros(Np, float)
    gh2 = g * h * h
    flag = True
    counter = 0
    divider = 10

    plt.subplot(2, 3, idx)
    plt.title(f"x = {x_val}")

    while flag:
        flag = False
        counter += 1

        for i in range(1, N):
            xp[i] = (x[i+1] + x[i-1] + gh2) * 0.5
            if flag == False:
                if abs(xp[i] - x[i]) > target:
                    flag = True

            x = 1.0 * xp
            if counter % divider == 0:
                plt.plot(np.arange(a, b+h, h), x)
                divider *= 10

    plt.plot(np.arange(a, b+h, h), x)
    plt.xlabel("time(s)")
    plt.ylabel("height(m)")

plt.tight_layout()
plt.show()
```

```python
#Shooting Method
import numpy as np
import matplotlib.pyplot as plt

g = 9.8
a = 0.0
b = 10.0
N = 1000
h = (b-a)/N
target = 1e-10

def RK4(f,y0,t,h0):
  yi = y0
  yt = []
  for ti in t :
    yt.append(yi*1.0)
    k1 = 0.5*h0*f(yi,ti)
    k2 = 0.5*h0*f(yi+k1,ti+h2)
    k3 = h0*f(yi+k2,ti+h2)
    k4 = h0*f(yi+k3,ti+h)
    yi += (2*k1+4*k2+2*k3+k4)/6
  return np.array(yt,float)

def free_fall(r,t):
  y = r[0]
  vy = r[1]
  dydt = vy
  dvdt = -g
  return np.array( [ dydt, dvdt], float)

def final_height(v):
  r = np.array([0.0,v],float)
  tpoints = np.arange(a, b+h, h)
  rpoints = RK4(free_fall, r, tpoints, h)
  return rpoints [-1][0]

v1, v2 = 0.01, 1000.0
h1 = final_height(v1)
h2 = final_height(v2)

c=0

while abs(h2-h1)>target:

  vp = (v1+v2)*0.5
  hp = final_height(vp)
  if h1*hp >0:
    v1 ,h1 = vp, hp
  else:
    v2, h2 = vp, hp
  c+=1


vf = (v1+v2)*0.5
hf = final_height(vf)

print(vf,hf)

r = np.array([0.0,vf],float)
tpoints = np.arange(a,b+h,h)
rpoints = RK4(free_fall, r, tpoints, h)

plt.subplot(211)
plt.plot(tpoints,rpoints[:,0])
plt.subplot(212)
plt.plot(tpoints, rpoints[:,1])




```
